Setup Dataset

```python
import pandas as pd, numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Mengatur tema global agar grafik terlihat rapi dan estetis
sns.set_theme(style='whitegrid', palette='Set2')

# Memuat dataset diamonds bawaan Seaborn
df_diamonds = sns.load_dataset('diamonds')

# Menampilkan informasi ringkas dataset
print(df_diamonds.head())
```

Distribusi Harga Berlian

```python
fig, ax = plt.subplots(figsize=(9, 5))

# Membuat histogram harga dengan kurva KDE
sns.histplot(data=df_diamonds, x='price', bins=30, kde=True, color='#028090', ax=ax)

# Menambahkan komponen anatomi chart yang baik
ax.set_title('Distribusi Harga Berlian (Dataset Diamonds)', fontsize=14, pad=15)
ax.set_xlabel('Harga (USD)', fontsize=11)
ax.set_ylabel('Jumlah/Frekuensi', fontsize=11)

plt.tight_layout()
plt.show()
```

Perbandingan Harga Berdasarkan Kualitas Potongan

```python
fig, ax = plt.subplots(figsize=(10, 6))

# Membuat Boxplot harga per kategori kualitas potongan (cut)
sns.boxplot(data=df_diamonds, x='cut', y='price', palette='Set2', ax=ax)

# Menambahkan komponen label dan judul
ax.set_title('Perbandingan Harga Berlian Berdasarkan Kualitas Potongan (Cut)', fontsize=14, pad=15)
ax.set_xlabel('Kualitas Potongan (Cut)', fontsize=11)
ax.set_ylabel('Harga (USD)', fontsize=11)

plt.tight_layout()
```

Hubungan antara Karat (Carat) dan Harga (Price)

```python
fig, ax = plt.subplots(figsize=(10, 6))

# Mengambil sampel data (sampling) sebanyak 2000 baris agar plot tidak terlalu padat dan berat
df_sample = df_diamonds.sample(n=2000, random_state=42)

# Membuat Scatter plot dengan dimensi ketiga (warna berdasarkan Clarity)
sns.scatterplot(data=df_sample, x='carat', y='price', hue='clarity', alpha=0.6, palette='viridis', ax=ax)

# Menambahkan komponen pendukung
ax.set_title('Hubungan antara Karat Berlian dan Harga (Sampel 2.000 Data)', fontsize=14, pad=15)
ax.set_xlabel('Karat (Carat)', fontsize=11)
ax.set_ylabel('Harga (USD)', fontsize=11)
ax.legend(title='Kejernihan (Clarity)', bbox_to_anchor=(1.05, 1), loc='upper left')

# Menghapus border atas dan kanan sesuai prinsip efisiensi
ax.spines[['top', 'right']].set_visible(False)

plt.tight_layout()
plt.show()
```
