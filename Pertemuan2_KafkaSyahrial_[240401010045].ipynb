Q1 Berapa jumlah baris dan kolom pada dataset Titanic? Kolom apa saja yang tersedia?

```python
print(df.shape)
print(df.columns)
```

Hasil: Jumlah kolom 12 dari 891 dataset

Q2 Kolom mana saja yang memiliki missing values Berapa jumlah dan proporsinya (%) per kolom?

```python
missing_value = df.isnull().sum()
persentase = (missing_value / len(df)) * 100

hasil_missing = pd.DataFrame({
      'Jumlah Missing': missing_value,
      'Persentase (%)': persentase
})

print(hasil_missing)
```

Hasil:
- Age > *missing: 177 Presentase: 19.865320%*
- Cabin > *missing: 687 Presentase: 77.104377%*
- Embarked > *missing: 687 Presentase: 77.104377%*

Q3 Berapa persen penumpang yang selamat (Survived=1)? Gunakan operasi NumPy atau Pandas.

```python
penumpang_selamat = df['Survived'].mean() * 100
print(f"Persentase penumpang yang selamat: {penumpang_selamat:.2f}%")
```

Hasil: Penumpang selamat berjumlah 38.38%

Q4 Filter penumpang wanita dari kelas 1 (Pclass=1). Berapa jumlahnya dan berapa rata-rata usianya?

```python
wanita_kls1 = df[(df['Sex'] == 'female') & (df['Pclass'] == 1)]

jumlah = len(wanita_kls1)
rata_usia = wanita_kls1['Age'].mean()

print("Jumlah Wanita Kelas 1:", jumlah)
print("Rata-rata Usia Wanita Kelas 1:", rata_usia)
```

Hasil: Wanita kelas 1 berjumlah 94 dengan rata-rata usia 34

Q5 Gunakan groupby untuk menghitung tingkat keselamatan (mean Survived) per kelas penumpang. Kelas mana yang paling tinggi?

```python
keselamatan = df.groupby('Pclass')['Survived'].mean()
print(keselamatan)
```

Hasil: Kelas 1

Q6 Buat analisis tambahan berdasarkan variabel lain yang Anda anggap menarik. Sertakan interpretasi singkat.

```python
rata_umur_selamat = df.groupby('Survived')['Age'].mean()
print(rata_umur_selamat)
```
